Apresente a soma de todos os argumentos inteiros ou reais recebidos

```java
import java.util.Scanner; //importação do Scanner para ler os valores digitados
Scanner scanner = new Scanner(System.in); //receber o que for digitado no teclado pelo usuário
        System.out.print("Digite os números separados por espaço: "); 
        String input = scanner.nextLine(); // lê a entrada do usuário e armazena na variável input como uma string
        String[] valores = input.split(" "); // separa a string input em um array de strings chamado "valores", usando o espaço como delimitador
        double soma = 0; // inicializa a variável soma como 0
        for (String valor : valores) { // itera sobre cada valor do array "valores"
            if (valor != null && !valor.trim().isEmpty()) { // verifica se o valor não é nulo ou vazio
                try { // tenta converter o valor para um número double
                    soma += Double.parseDouble(valor);  // adiciona o valor à soma
                } catch (NumberFormatException e) { // caso ocorra uma exceção de conversão de tipo (NumberFormatException)
                     // Argumento inválido: desconsiderar (não faz nada)
                }
            }
        }
        System.out.println("A soma dos números informados é: " + soma); // imprime na tela a mensagem "A soma dos números informados é: " seguida da soma dos números informados
```
