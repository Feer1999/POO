Mostre a média usando o showMessageDialog da classe JOptionPane

```java
import javax.swing.JOptionPane;´//importa a classe JOptionPane da biblioteca swing.
String nota1 = JOptionPane.showInputDialog("Digite a nota da primeira prova:"); // exibe uma caixa de diálogo pedindo ao usuário a nota da primeira prova e armazena a entrada em uma variável do tipo String chamada nota1
double prova1 = Double.parseDouble(nota1);//converte a entrada do usuário (que está em formato de String) em um valor numérico do tipo double e armazena na variável prova1.
String nota2 = JOptionPane.showInputDialog("Digite a nota da segunda prova:"); //exibe uma caixa de diálogo pedindo ao usuário a nota da segunda prova e armazena a entrada em uma variável do tipo String chamada nota2.
double prova2 = Double.parseDouble(nota2);//converte a entrada do usuário (que está em formato de String) em um valor numérico do tipo double e armazena na variável prova2.
String notaTrabalho = JOptionPane.showInputDialog("Digite a nota do trabalho:");// exibe uma caixa de diálogo pedindo ao usuário a nota do trabalho e armazena a entrada em uma variável do tipo String chamada notaTrabalho.
double trabalho = Double.parseDouble(notaTrabalho);//converte a entrada do usuário (que está em formato de String) em um valor numérico do tipo double e armazena na variável trabalho.
double media = (prova1 + prova2 + trabalho) / 3; // calcula a média das notas e armazena o resultado na variável media.
JOptionPane.showMessageDialog(null, "A média das notas é: " + media);//exibe uma caixa de diálogo com a mensagem "A média das notas é: " seguida do valor da variável media.
```
